```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread('img/sungchul.jpg')
print(type(img))
print(img.shape)
print(img.dtype)
plt.imshow(img)
plt.show()
```

```python
img_flipped = cv2.flip(img, 1) # 1 - left_right_flip
print(type(img_flipped))
print(img_flipped.shape)
print(img_flipped.dtype)
plt.imshow(img_flipped)
plt.show()
```

```python
img_flipped = img[:,::-1]
print(type(img_flipped))
print(img_flipped.shape)
print(img_flipped.dtype)
plt.imshow(img_flipped)
plt.show()
```

```python
img_flipped = cv2.flip(img, 0) # 0 - up_down_flip
print(type(img_flipped))
print(img_flipped.shape)
print(img_flipped.dtype)
plt.imshow(img_flipped)
plt.show()
```

```python
img_flipped = img[::-1,:,:]
print(type(img_flipped))
print(img_flipped.shape)
print(img_flipped.dtype)
plt.imshow(img_flipped)
plt.show()
```
