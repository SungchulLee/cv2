```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread('img/sungchul.jpg')
print(type(img))
print(img.shape)
print(img.dtype)
plt.imshow(img)
plt.show()
```

```python
dsize = (100, 200)
interpolation = cv2.INTER_NEAREST
img = cv2.resize(img, dsize=dsize, interpolation=interpolation)
print(type(img))
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
height, width = img.shape[:2]
center = (width/2, height/2)
angle = -45 
scale = 2 
M = cv2.getRotationMatrix2D(center, angle, scale)
img = cv2.warpAffine(img, M, (width, height))
print(type(img))
print(img.shape)
plt.imshow(img)
plt.show()
```
