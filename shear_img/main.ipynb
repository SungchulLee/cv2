```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread('img/sungchul.jpg')
img = cv2.resize(img, dsize=(200,100), interpolation=cv2.INTER_NEAREST)
print(type(img))
print(img.shape)
print(img.dtype)
plt.imshow(img)
plt.show()
```

```python
height, width, _ = img.shape
base = np.float32([0, height])
right = np.float32([width, height])
up = np.float32([0, 0])

shear_range = 0.2
shear = shear_range*(2*np.random.uniform()-1)
new_up = np.float32([width*shear, 0])

points_before = np.float32([base,right,up])
points_after = np.float32([base,right,new_up])
M = cv2.getAffineTransform(points_before, points_after)

img_sheared = cv2.warpAffine(img, M, (width, height))
print(type(img_sheared))
print(img_sheared.shape)
plt.imshow(img_sheared)
plt.show()
```
