```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread('img/sungchul.jpg')
print(type(img))
print(img.shape)
print(img.dtype)
plt.imshow(img)
plt.show()
```

```python
dsize = (100, 200)
#dsize = (200, 100)
#dsize = (200, 200)

#interpolation = cv2.INTER_CUBIC 
interpolation = cv2.INTER_NEAREST

img = cv2.resize(img, dsize=dsize, interpolation=interpolation)
print(type(img))
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
height, width = img.shape[:2]

t_x = 0
#t_x = 50
#t_x = -50

#t_y = 0 
t_y = 50
#t_y = -50

M = np.float32([ [1,0,t_x], [0,1,t_y] ])

new_width = width 
#new_width = int(1.5*width)

#new_height = height 
new_height = int(1.5*height)

img_translate = cv2.warpAffine(img, M, (new_width, new_height))
plt.imshow(img_translate)
plt.show()
```
