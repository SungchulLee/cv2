```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread('img/sungchul.jpg')
img = cv2.resize(img, dsize=(200,200), interpolation=cv2.INTER_NEAREST)
print(type(img))
print(img.shape)
print(img.dtype)
plt.imshow(img)
plt.show()
```

```python
height, width, _ = img.shape
zoom_range = 0.5 # keras convention (https://keras.io/preprocessing/image/)
zoom_range = 1 + zoom_range
crop_range = 1 / zoom_range
crop_factor = crop_range + (1-crop_range)*np.random.uniform() 
crop_height = int(height * crop_factor)
crop_width = int(width * crop_factor)

max_width_start = width - crop_width
max_height_start = height - crop_height
width_start = np.random.randint(0, max_width_start)
height_start = np.random.randint(0, max_height_start)

img_clopped = img[width_start:width_start+crop_width, 
                  height_start:height_start+crop_height]
img_clopped = cv2.resize(img_clopped, dsize=(200,200), interpolation=cv2.INTER_NEAREST)
print(type(img_clopped))
print(img_clopped.shape)
plt.imshow(img_clopped)
plt.show()
```
